```python
# p = Path('../src/data/lab.val')
# out_p = Path('../src/data/lab_pose.val')

p = Path('../src/data/explorationroom_largest.val')
out_p = Path('../src/data/explorationroom_largest-pose.val')
if not out_p.is_file():
    out_p.touch()
out = out_p.open("a")
```

### walking,sitting,standing2legs,standing4legs,NiS

```python
i = 2992
with p.open('r') as f:
    for j, img in enumerate([x.rstrip() for x in f.readlines()]):
        if j < i:
            continue

        img = cv2.imread(img)
        while(1):
            cv2.imshow('Image Pose?', img)
            k = cv2.waitKey(33)
            if k==27:    # Esc key to stop
                print('stopped after: ', str(i))
                exit()
            elif k==-1:  # normally -1 returned,so don't print it
                continue
            elif k>=ord('1') and k<=ord('5'):
                out.write(chr(k-1)+'\n')
                break
        i += 1
```
